```python
import numpy as np

mat = (np.random.random((12,11))*25).round().astype(np.uint)
words = [("a", (1,1), True),
("a", (2,0), False),
("a", (10,0), True)]

for i, j, h in words:
    word = i.upper()
    x, y = j
    n = np.array([[ord(i)-65 for i in word]])
    if(not h):
        n = n.transpose()
        mat[y:y+len(word),x:x+1]=n
    else:
        mat[x:x+1,y:y+len(word)]=n
a = ""
for i in mat:
    a += " & ".join([chr(j+65) for j in i])+" \\\\\n"
    
print(a)
```

```python
import re
form = []
with open("./form-prof.md", "r") as file:
    line1 = file.readline()
    while line1 != "":
        line1=line1.strip()
        if re.search("^[0-9]*[.]",line1):
            form.append(f"{int(len(form)/2+1)}. " + re.sub("^[0-9]*[.]", " ",line1).strip())
            form.append(file.readline().strip())
            print(form[-2])
        line1 = file.readline()
```

```python
res = ""
for i in range(0,len(form), 2):
    op = form[i+1]
    index = int(1+i/2)
    res+="<div class=\"field col\">\n"+ form[i]+"\n"
    op = op.strip()
    if op == "range":
        res+= "<div class=\"range-box\"><input type=\"range\" value=\"0\" "
        res+= "id=\"q"+str(index)+"\" min=\"1\" max=\"5\" name=\"q"+str(index)+"\" list=\"values\" />"
        res+="""<div class="values-box"><p>1</p><p>2</p><p>3</p><p>4</p><p>5</p></div></div>"""
    elif op == "text":
        res+= f"<textarea data-optional=\"true\" name=\"q{index}\" id=\"q{index}\"></textarea>"
    elif op.startswith("{}"):
        res+="<div class=\"field-grid\">"
        op = [x.strip() for x in op.split("{}") if x.strip() != ""]
        
        for i, x in zip(range(len(op)),op):
            res+= "<div>\n"
            res+= f"<input type=\"checkbox\" id=\"q{i}\" name=\"q{index}\""
            res+= f"value=\"{x}\" />"
            res+= f"<label for=\"q{i}\">"
            res+= x+"</label></div>"
        res+= "</div>"
    else:
        res+="<div class=\"col\">"
        op = [x.strip() for x in op.split("[]") if x.strip() != ""]
        
        for i, x in zip(range(len(op)), op):
            res+= "<div>\n"
            res+= f"<input type=\"radio\" id=\"q{i}\" name=\"q{index}\""
            res+= "value=\"true\" />"
            res+= f"<label for=\"q{i}\">"
            res+= x+"</label></div>"
        res+= "</div>"
    
    res+="</div>\n<hr />"

print(res)
```
