```python
import numpy as np

mat = (np.random.random((12,11))*25).round().astype(np.uint)
words = [("a", (1,1), True),
("a", (2,0), False),
("a", (10,0), True)]

for i, j, h in words:
    word = i.upper()
    x, y = j
    n = np.array([[ord(i)-65 for i in word]])
    if(not h):
        n = n.transpose()
        mat[y:y+len(word),x:x+1]=n
    else:
        mat[x:x+1,y:y+len(word)]=n
a = ""
for i in mat:
    a += " & ".join([chr(j+65) for j in i])+" \\\\\n"
    
print(a)
```

```python
import re
res = []
with open("./formulario.md", "r") as file:
    line1 = file.readline()
    while line1 != "":
        line1=line1.strip()
        if re.search("^[0-9]*[.]",line1):
            res.append(f"{int(len(res)/2+1)}. " + re.sub("^[0-9]*[.]", " ",line1).strip())
            res.append(file.readline().strip())
            print(res[-2])
        line1 = file.readline()
```

```python
b = res
res = ""
for i in range(0,len(b), 2):
    op = b[i+1]
    index = int(1+i/2)
    res+="<div class=\"field col\">\n"+ b[i]+"\n"
    op = op.strip()
    if op == "range":
        res+= "<div class=\"range-box\"><input type=\"range\" value=\"0\" "
        res+= "id=\"q"+str(index)+"\" min=\"1\" max=\"5\" name=\"q"+str(index)+"\" list=\"values\" />"
        res+="""<div class="values-box"><p>1</p><p>2</p><p>3</p><p>4</p><p>5</p></div></div>"""
    elif op == "text":
        res+= f"<textarea data-optional=\"true\" name=\"q{index}\" id=\"q{index}\"></textarea>"
    else:
        res+="<div class=\"row\">"
        op = [x.strip() for x in op.split("[]") if x.strip() != ""]
        
        for x in op:
            res+= "<div>\n"
            res+= "<input type=\"radio\" name=\"q"+str(index)+"\""
            res+= "value=\""+ x +"\" />"
            res+= "<label for=\"q" + str(index)+ "\">"
            res+= x+"</label></div>"
        res+= "</div>"
    
    res+="</div>\n<hr />"

print(res)
```
